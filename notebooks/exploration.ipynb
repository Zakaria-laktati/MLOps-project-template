# Exploration Notebook

This notebook is used for exploratory data analysis (EDA) to understand the dataset and visualize data distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = '../data/raw/dataset.csv'  # Update with your dataset path
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Visualize data distributions
plt.figure(figsize=(12, 6))
sns.histplot(df['column_name'], bins=30, kde=True)  # Replace 'column_name' with your column
plt.title('Distribution of Column Name')
plt.xlabel('Column Name')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
